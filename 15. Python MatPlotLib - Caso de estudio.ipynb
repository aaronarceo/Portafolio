# Ejercicio 2: Incendios en México

```python
df_incendios= pd.read_csv("datos/Serie_historica_anual_incendios_2017.csv")
```

```python
# Visualizar los primeros 5 datos
df_incendios.head()
```

```python
# Datos principales
df_incendios.describe() # Los datos están completos en todas las columnas
```

<div class="alert alert-success">
    
1. Utilizar una grafica de barras para comparar el total de hectáreas afectadas para los 5 estados más afectados 
    
</div> 

```python
# Definir los estados con más hectáreas afectadas
valores_estados = df_incendios.groupby(['Estado'])['Total Hectareas'].sum().sort_values(ascending=False).head().index
valores_hectareas = df_incendios.groupby(['Estado'])['Total Hectareas'].sum().sort_values(ascending=False).head().values
```

```python
# Estilo MATLAB
plt.figure(figsize=(10,4))

plt.bar(valores_estados, valores_hectareas, facecolor='blue', width=0.6)
plt.title('Total de hectáreas afectadas por estado')
plt.xlabel('Estados')
plt.ylabel('Total de hectáreas afectadas')
plt.show()
```

```python
# Estilo orientado a objetos
fig, ax = plt.subplots(figsize=(10,4))

ax.bar(valores_estados, valores_hectareas, facecolor = 'blue', width=0.6)
ax.set(title='Total de hectáreas afectadas por estado', xlabel='Estados', ylabel='Total de hectáreas afectadas')

plt.show()
```

<div class="alert alert-success">
    
2. Utilizar una grafica de barras para comparar el total de hectáreas afectadas para Yucatán para los meses: abril, mayo, junio, julio y agosto 
    
</div> 

```python
# Data frame de Yucatán
df_inc_yuc = df_incendios[df_incendios['Estado'] == 'Yucatán']

# Total de hectareas afectadas por mes en Yucatán
df_inc_yuc.groupby(['Mes Acontecimiento'])['Total Hectareas'].sum().values

# Asignar valores
    #valores_meses = df_inc_yuc.groupby(['Mes Acontecimiento'])['Total Hectareas'].sum().index
    #valores_hectareas_yuc = df_inc_yuc.groupby(['Mes Acontecimiento'])['Total Hectareas'].sum().values
valores_meses = df_inc_yuc.groupby(['Mes Acontecimiento'])['Total Hectareas'].sum().loc[['abril', 'mayo', 'agosto']].index
valores_hectareas_yuc = df_inc_yuc.groupby(['Mes Acontecimiento'])['Total Hectareas'].sum().loc[['abril', 'mayo', 'agosto']].values
```

```python
# Estilo MATLAB
plt.figure(figsize=(4,4))

plt.bar(valores_meses, valores_hectareas_yuc, facecolor='green', width=0.5)
plt.title('Total de hectáreas afectadas para yucatán')
plt.xlabel('Meses')
plt.ylabel('Total de Hectáreas afectadas')

plt.show()
```

```python
# Estilo orientado a objetos
fig, ax = plt.subplots(figsize=(4,4))

ax.bar(valores_meses, valores_hectareas_yuc, facecolor='green', width=0.5)
ax.set(title='Total de hectáreas afectadas para yucatán', xlabel='Meses', ylabel='Total de Hectáreas afectadas')

plt.show()
```

# Ejercicio 3: Pacientes confirmados con Covid en México


Fuente: https://datos.covid-19.conacyt.mx/#DownZCSV

```python
df_covid = pd.read_csv("datos/Casos_Diarios_Estado_Nacional_Confirmados_20201129.csv")
df_covid.head()
df_covid = df_covid.loc[:,'nombre':].set_index('nombre').T
df_covid.head()
```

```python
df_covid = df_covid.reset_index().rename(columns = {'index': 'Fecha'})
df_covid.head()
```

```python
df_covid['Fecha'] = pd.to_datetime(df_covid['Fecha'], format = '%d-%m-%Y')
```

```python
df_covid = df_covid.set_index('Fecha')
```

```python
df_covid
```

<div class="alert alert-success">
    
1. Hacer una grafica de barras comparando todos los meses, considerando el total nacional (por mes) 
    
</div> 

```python
# Datos mensuales
df_covid.resample('M')['Nacional'].sum()
```

```python
# Asignación de valores
valores_cvd_meses = df_covid.resample('M')['Nacional'].sum().index
valores_cvd_nacional = df_covid.resample('M')['Nacional'].sum().values
```

```python
# Estilo MATLAB
plt.figure(figsize=(12,4))

meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre' ]
# plt.bar(valores_cvd_meses, valores_cvd_nacional, width=10)
plt.bar(meses, valores_cvd_nacional)
plt.title('Total de casos por mes a nivel nacional')
plt.xlabel('Meses')
plt.ylabel('Cantidad de contagios')

plt.show()
```

```python
# Estilo orientado a objetos

fig, ax = plt.subplots(figsize=(12,4))

ax.bar(meses, valores_cvd_nacional)
ax.set(title='Total del casos por mes a nivel nacional', xlabel = 'Meses', ylabel = 'Cantidad de contagios')

plt.show()
```

<div class="alert alert-success">
    
2. Hacer una grafica de barras comparando  los meses de agosto, septiembre, octubre y noviembre, considerando solo las entidades de Yucatán, Distrito Federal  y Jalisco
    
</div> 

```python
# Datos mensuales por estado
df_covid[['YUCATAN', 'DISTRITO FEDERAL','JALISCO']].resample('M').sum().loc['2020-07-31':]
```

```python
# Asignación de valores
valores_yuc = df_covid['YUCATAN'].resample('M').sum().values[7:11]
valores_cdmx = df_covid['DISTRITO FEDERAL'].resample('M').sum().values[7:11]
valores_jal = df_covid['JALISCO'].resample('M').sum().values[7:11]
```

```python
x_vals = np.arange(len(meses[7:11]))
ancho = 0.25
```

```python
# Estilo MATLAB
plt.figure(figsize=(8,4))

plt.bar(x_vals - ancho, valores_yuc, width= ancho, label='YUC')
plt.bar(x_vals, valores_cdmx, width=ancho, label='CDMX')
plt.bar(x_vals + ancho, valores_jal, width=ancho, label='JAL')

plt.xticks(x_vals, meses[7:11])
plt.title('Casos de covid por mes y estados')
plt.xlabel('Meses')
plt.ylabel('Casos de covid confirmados')
plt.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,4))

ax.bar(x_vals - ancho, valores_yuc, width= ancho, label='YUC')
ax.bar(x_vals, valores_cdmx, width=ancho, label='CDMX')
ax.bar(x_vals + ancho, valores_jal, width=ancho, label='JAL')

ax.set_xticks(x_vals)
ax.set_xticklabels(meses[7:11])
ax.legend()
ax.set(title= 'Casos de covid por mes y estados', xlabel = 'Meses', ylabel = 'Casos de covid confirmados')

plt.show()
```

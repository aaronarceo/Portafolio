```python
# Número total de hombres y mujeres
contador_hombres=0
contador_mujeres=0
contador_sobrevivientes=0
contador_sobrevivientes_mujeres=0
with open('datos/titanic_datos.txt') as csvarchivo:
    lector = csv.reader(csvarchivo)
    next(lector)
    for linea in lector:
        if ('female' in linea[5]):
            contador_mujeres = contador_mujeres + 1
        else:
            contador_hombres = contador_hombres + 1
        if (linea[1]=='1'):
            contador_sobrevivientes = contador_sobrevivientes + 1
        if (linea[1]=='1' and linea[5]=='female'):
            contador_sobrevivientes_mujeres = contador_sobrevivientes_mujeres + 1

# Contador de hombres y mujeres
print('Total de hombre:', contador_hombres)
print('Total de mujeres:', contador_mujeres)
contador_personas = contador_hombres + contador_mujeres

#print('Total del pasajeros:', contador_personas)
print('Probabilidad de supervivencia de una personas:', "{0:.2f}".format(100*(contador_sobrevivientes/contador_personas)),"%")

#print('Sobrevivientes mujeres:', contador_sobrevivientes_mujeres)

#print('contador de sobrevivientes:', contador_sobrevivientes)
contador_sobrevivientes_hombres = contador_sobrevivientes - contador_sobrevivientes_mujeres
print('Probabilidad de supervivencia de hombres:', "{0:.2f}".format(100*(contador_sobrevivientes_hombres/contador_hombres)),"%")
print('Probabilidad de supervivencia de mujeres:', "{0:.2f}".format(100*(contador_sobrevivientes_mujeres/contador_mujeres)),"%")
```

```python
edad_promedio=0
count_promedio=0
edades = []
with open('datos/titanic_datos.txt') as csvarchivo:
    lector = csv.reader(csvarchivo)
    next(lector)
    for linea in lector:
        try:
            edad_promedio = edad_promedio + int(linea[6])
            count_promedio = count_promedio + 1
            edades.append(int(linea[6]))
        except:
            pass
    cinco_edades=[]
    print('La edad promedio es:', "{0:.2f}".format(edad_promedio/count_promedio))
    print('La edad de la persona más grande es:', max(edades))
    print('La edad de la persona más pequeña es:', min(edades))
    
    for i in range (0,5):
        max_edad = 0
        for j in range(0,len(edades)):
            if edades[j]>max_edad:
                max_edad=edades[j]  
        edades.remove(max_edad)
        cinco_edades.append(max_edad)
        
print('Las cinco edades más altas:',cinco_edades)

# print(edades)
```

  <li> 6. ¿Cuál hubiera sido la ganancia total? Considere que el día del embarque se le pedía un impuesto a cada tripulante dependiendo de su clase (3era = 0.5 pounds, 2da = 15 pounds, 1era=40 pounds).  Defina una función llamada conversión que reciba dos parámetros (la clase y el costo del ticket) y  regrese el costo actual (USD). Exprese su resultados en USD  </li>

```python
count_c1=0
count_c2=0
count_c3=0
tax=0
with open('datos/titanic_datos.txt') as csvarchivo:
    lector = csv.reader(csvarchivo)
    next(lector)
    for linea in lector:
        tax = tax + float(linea[10])
        if linea[2] == '1':
            count_c1 = count_c1 + 1
        if linea[2] == '2':
            count_c2 = count_c2 + 1
        if linea[2] == '3':
            count_c3 = count_c3 + 1
                    
print('Número de pasajeros de clase 1:', count_c1)
print('Número de pasajeros de clase 2:', count_c2)
print('Número de pasajeros de clase 3:', count_c3)


def boleto(costo,clase):
    if clase == 'c1':
        costo_mas_imp = (costo + 40)*1.18
    if clase == 'c2':
        costo_mas_imp = (costo + 15)*1.18
    if clase == 'c3':
        costo_mas_imp = (costo + 0.40)*1.18
    return print('El costo de su boleto es:',costo_mas_imp,'Dólares')

ganancia = count_c1*40 + count_c2*15 + count_c3*0.5 + tax
print('La ganacia total del titanic sería:', ganancia*1.18, 'Dólares\n')

clase = input('¿Cuál es su clase (c1, c2 o c3)? ')
costo = int(input('¿Cuál es el costo del boleto en libras: '))
boleto(costo,clase)
    
```

  <li> 8. ¿Cómo difiere la tasa de supervivencia entre las diferentes clases del Titanic?</li>

```python
count_sobreviviente_c1=0
count_sobreviviente_c2=0
count_sobreviviente_c3=0

with open('datos/titanic_datos.txt') as csvarchivo:
    lector = csv.reader(csvarchivo)
    next(lector)
    for linea in lector:
        if linea[2] == '1' and linea[1] == '1':
            count_sobreviviente_c1 += 1
        if linea[2] == '2' and linea[1] == '1':
            count_sobreviviente_c2 += 1
        if linea[2] == '3' and linea[1] == '1':
            count_sobreviviente_c3 += 1
print('La probabilidad de supervivencia de la Clase 1 es:', '{0:.2f}'.format(100*(count_sobreviviente_c1/count_c1)),'%')
print('La probabilidad de supervivencia de la Clase 2 es:', '{0:.2f}'.format(100*(count_sobreviviente_c2/count_c2)),'%')
print('La probabilidad de supervivencia de la Clase 3 es:', '{0:.2f}'.format(100*(count_sobreviviente_c3/count_c3)),'%')
```
